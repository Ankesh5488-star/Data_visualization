```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
border_color = 'w'  # Border color
border_width = 3      # Border width
# Example data
labels = ['Artificial Inteligence', 'Machine Learning', 'Data science', 'Deep Learning']
sizes = [20, 30, 20, 25]
colors = ['blue', 'green', 'orange', 'red']

plt.figure(figsize=(8, 8))
plt.title("Simple Pie Chart",fontsize=30,color='w',fontweight='bold')

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=100,wedgeprops={'edgecolor': border_color, 'linewidth': border_width},textprops={'color':'w','fontsize':12})

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.legend(loc='center',bbox_to_anchor=(1.3, 0.5))
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np

# Generate data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
# ax.scatter3D(X, Y, Z, c=Z, cmap='viridis')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot Example')

ax.view_init(elev=20, azim=45)

plt.show()
```

```python
x=np.linspace(0,10,100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x)

y_random = np.random.rand(100)
plt.figure(figsize=(10,6))
plt.plot(x, y_sin,'b', label='Sine Wave')
plt.plot(x, y_cos,'r', label='Cosine Wave')
intersection_indices = np.where(np.isclose(y_sin,y_cos,atol=0.05))
plt.scatter(x[intersection_indices], y_sin[intersection_indices], color='k', marker='o', label='Intersections')
plt.xlabel ('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.fill_between(x,y_sin,y_cos,color='gray',alpha=0.3,label="filled area")
plt.legend(loc='upper right',bbox_to_anchor=(1.2,1))
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (Number of Hours Worked vs. Payment for Work)
hours_worked = [10,10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
payment = [100,120, 150, 200, 250, 300, 350, 400, 450, 500, 550]

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(hours_worked, payment, color='blue', label='Data Points')

# Add labels and title
plt.xlabel('Number of Hours Worked')
plt.ylabel('Payment for Work (in $)')
plt.title('Scatter Plot: Hours Worked vs. Payment for Work')

coefficients = np.polyfit(hours_worked, payment, 1)
trendline = np.poly1d(coefficients)
plt.plot(hours_worked, trendline(hours_worked), color='red', label='Trendline (Linear Regression)')

# Add a legend
plt.legend(loc='upper right',bbox_to_anchor=(1.45,1))

# Show the plot
plt.grid(True)
plt.show()
```
